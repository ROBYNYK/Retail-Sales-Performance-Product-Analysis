```python
# 1. Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings("ignore")
```

```python
# 2. Load the dataset from Documents folder
Sales_Coffee = r"C:\Users\Rob\Documents\Coffee Shop Sales.xlsx"
df = pd.read_excel(Sales_Coffee)
# Preview dataset shape and first rows
print("Data loaded successfully! Shape:", df.shape)
print(df.head())
```

```python
# 4. Clean column names (strip spaces)
df.columns = df.columns.str.strip()
```

```python
# 5. Convert transaction_date to datetime
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
df['transaction_time'] = pd.to_datetime(df['transaction_time'], format='%H:%M:%S').dt.time

# Extract useful time-based features
df['hour'] = pd.to_datetime(df['transaction_time'].astype(str)).dt.hour
df['day_of_week'] = df['transaction_date'].dt.day_name()

# 6. Create a revenue column
df['revenue'] = df['transaction_qty'] * df['unit_price']
```

```python
# Exploratory Analysis


# Orders by Hour of Day ---
orders_by_hour = df.groupby('hour')['transaction_id'].count()
plt.figure(figsize=(10,6))
orders_by_hour.plot(kind='bar', color='skyblue')
plt.title("Orders by Hour of Day")
plt.xlabel("Hour of Day (24h format)")
plt.ylabel("Number of Orders")
plt.xticks(rotation=0)
plt.show()

# Orders by Day of Week ---
orders_by_day = df.groupby('day_of_week')['transaction_id'].count().reindex(
    ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
)
plt.figure(figsize=(8,5))
orders_by_day.plot(kind='bar', color='orange')
plt.title("Orders by Day of Week")
plt.xlabel("Day")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)
plt.show()

# Sales by Product Category ---
sales_by_category = df.groupby('product_category')['unit_price'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sales_by_category.plot(kind='bar', color='green')
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Sales Revenue")
plt.xticks(rotation=45)
plt.show()

# Top 10 Best-Selling Items ---
top_items = df.groupby('product_detail')['transaction_id'].count().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
top_items.plot(kind='bar', color='purple')
plt.title("Top 10 Most Ordered Items")
plt.xlabel("Product")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45, ha='right')
plt.show()

# Revenue Trend Over Time ---
daily_sales = df.groupby('transaction_date')['unit_price'].sum()
plt.figure(figsize=(12,6))
daily_sales.plot(color='blue')
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()

# Heatmap: Orders by Day of Week & Hour
pivot_table = df.pivot_table(
    index='day_of_week',
    columns='hour',
    values='transaction_id',
    aggfunc='count'
).reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])

plt.figure(figsize=(12,6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".0f")
plt.title("Heatmap of Orders (Day of Week vs Hour)")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.show()

# Store location distribution 
store_location_counts = df['store_location'].value_counts()
plt.figure(figsize=(8,8))
store_location_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap="Set3")
plt.title("Store Location Distribution")
plt.ylabel("")
plt.show()
```

```python
# Time Series Analysis
from statsmodels.tsa.stattools import adfuller  # Import the adfuller function
from statsmodels.tsa.statespace.sarimax import SARIMAX  # Also import SARIMAX since it's used later
import matplotlib.pyplot as plt  # Import matplotlib for plotting
import pandas as pd  # Import pandas for date handling

# Daily revenue
daily_rev = df.groupby('transaction_date')['revenue'].sum()

# Check for stationarity
adf_test = adfuller(daily_rev)
print("\nADF Statistic:", adf_test[0])
print("p-value:", adf_test[1])

# Differencing if not stationary
if adf_test[1] > 0.05:
    daily_rev_diff = daily_rev.diff().dropna()
else:
    daily_rev_diff = daily_rev

# Fit SARIMAX model
model = SARIMAX(daily_rev, order=(1,1,1), seasonal_order=(1,1,1,7))  # weekly seasonality assumed
results = model.fit(disp=False)

# Forecast next 30 days
forecast = results.get_forecast(steps=30)
pred_mean = forecast.predicted_mean
pred_ci = forecast.conf_int()


# Plot historical and forecast revenue
plt.figure(figsize=(12,6))
plt.plot(daily_rev.index, daily_rev.values, label="Historical Revenue")
plt.plot(pred_mean.index, pred_mean.values, label="Forecasted Revenue", color="orange")
plt.fill_between(pred_ci.index,
                 pred_ci.iloc[:,0],
                 pred_ci.iloc[:,1],
                 color="orange", alpha=0.2)
plt.title("Revenue Forecast (Next 30 Days)")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.legend()
plt.show()

# Save forecast to DataFrame for further analysis if needed
forecast_df = pd.DataFrame({
    'date': pred_mean.index,
    'forecasted_revenue': pred_mean.values,
    'lower_ci': pred_ci.iloc[:,0].values,
    'upper_ci': pred_ci.iloc[:,1].values
})
print("\nForecast DataFrame:")
print(forecast_df)
```
